```python
def filter4uniq( table ):
    keeps=[]
    for i in range(len(table['query'])):
        s = list(table['subject'])[i]
        temp_table = (table[table['subject']==s]).sort_values(by=['%identity'],ascending=False)
        keeps.append(temp_table.index[0])
    keeps = list(set(keeps))
    return (table.iloc[keeps]).reset_index().drop(['index'],axis=1)

def record2dict(in_path):
    out={}
    temp_records = SeqIO.parse( in_path , "fasta")
    for seq_record in temp_records:
        ID = str(seq_record.id).strip()
        seq = str(seq_record.seq).strip()
        out[ID]=seq
    return out 
```

```python
def blast2table(file_path):    
    fields = ['query', 'subject', '%identity', 'alignment length', 'mismatches', 'gap opens', 'q. start', 'q. end',
              's. start', 's. end', 'evalue', 'bit score']
    blast_data = []
    dorecord = False
    with open(file_path) as f:
        lines = f.readlines()
        for l in lines:
            if '#' in l:
                if 'hits found' in l:
                    num = int(l.replace('#','').replace('hits found','').strip() )
                    if num == 0:
                        continue
                    else:
                        dorecord = True
                continue
            elif not dorecord:
                continue
            else:
                record = l.strip().split('\t')
                temp = {}
                for i in range(len(record)):
                    if i in [3,4,5,6,7,8,9]:
                        temp[fields[i]] = int(record[i])
                    elif i in [2,10,11]:
                        temp[fields[i]] = float(record[i])
                    else:
                        temp[fields[i]] = record[i]
                blast_data.append(temp)
                dorecord = False
    return pd.DataFrame(blast_data)
```

```python
def save_clean_fasta( in_path, out_path, sample ):
    temp_records = SeqIO.parse( in_path , "fasta")
    new_records = []
    for seq_record in temp_records:
        ID = sample + '_' + str(seq_record.id).split('#')[0].strip()
        seq = str(seq_record.seq)
        new_records.append( SeqRecord(Seq(seq), id = ID, name="",description="") )
    SeqIO.write(new_records, out_path ,"fasta")
    print( len(new_records) )
```

```python
kegg_tables={'s1b1':None,'s1b2':None,'s2b1':None,'s2b2':None,'s3b1':None,'s3b2':None}
for k in kegg_tables.keys():
    temp = pd.read_csv('../data/kegg_anno/'+k+'.KEGG.anno.csv')
    temp = (temp[temp['KO_PATHWAY']!= '-'] ).reset_index().drop(['index'],axis=1)
    kegg_tables[k] = temp[['Query_id','KO_ID','KO_DEFINITION','KO_EC','KO_PATHWAY']]

class_names = ['Metabolism','Genetic Information Processing','Cellular Processes',\
               'Environmental Information Processing']
```

```python
def get_pathway_anno( species, sample, query ):
    name = query
    query = query.replace('s1_','').replace('s2_','').replace('s3_','')
    if sample == 's1':
        if species == 'st':
            key = 's1b2'
        else:
            key = 's1b1'
    else:
        if species == 'st':
            key = sample+'b1'
        else:
            key = sample+'b2'
    table = kegg_tables[key]
    if query not in list(table['Query_id']):
        return 'None'
    ko_id = list((table[table['Query_id']==query])['KO_ID'])[0]
    path_info = list((table[table['Query_id']==query])['KO_PATHWAY'])[0]
    info_split = path_info.split('|')
    anno_sets = []
    for path in info_split:
        annos = path.split(';')
        if annos[1].strip() not in class_names:
            continue
        anno_sets.append({'query':name,'ko':ko_id,'class':annos[1].strip(),'subclass':annos[2].strip(),\
                          'pathway':annos[3].strip()})
    return anno_sets
```

```python
st_nr = record2dict('../data/MAGs/st_all/st_nr.faa')
lb_nr = record2dict('../data/MAGs/lb_all/lb_nr.faa')
```

```python
def sum_path(table):
    counts = {}
    genes = set(list(table['query']))
    for g in genes:
        temp_table = table[table['query'] == g]
        paths = set(list(temp_table['subclass']))
        for p in paths:
            if p not in counts:
                counts[p] = 1
            else:
                counts[p] += 1
    
    return counts
```

```python
cazy_tables={'s1b1':None,'s1b2':None,'s2b1':None,'s2b2':None,'s3b1':None,'s3b2':None}
for k in cazy_tables.keys():
    temp = pd.read_csv('../data/cazy_anno/'+k+'.anno.csv')
    kegg_tables[k] = temp
cazy_family_list = ['GH','GT','CBM','CE']
```

```python
def get_cazy_anno( species, sample, query ):
    name = query
    query = query.replace('s1_','').replace('s2_','').replace('s3_','')
    if sample == 's1':
        if species == 'st':
            key = 's1b2'
        else:
            key = 's1b1'
    else:
        if species == 'st':
            key = sample+'b1'
        else:
            key = sample+'b2'
    table = kegg_tables[key]
    if query not in list(table['Query_id']):
        return 'None'
    cazy_type = list((table[table['Query_id']==query])['CAZy_Family'])[0]
    for ca in cazy_family_list:
        if ca in cazy_type:
            ca_family = ca
    
    result = {'query':name,'family':ca_family,'group':cazy_type}
    
    return result
```

```python
st_annos,lb_annos = [],[]
for k in st_nr:
    sample = k.split('_')[0].strip()
    result = get_cazy_anno( 'st', sample, k )
    if result == 'None':
        continue
    st_annos.append(result)
for k in lb_nr:
    sample = k.split('_')[0].strip()
    result = get_cazy_anno( 'lb', sample, k )
    if result == 'None':
        continue
    lb_annos.append(result)

st_anno_table = pd.DataFrame(st_annos)
lb_anno_table = pd.DataFrame(lb_annos)
```

```python
groups = [ 'GT1','GT2','GT4','GT8', 'GT14','GT26','GT28','GT30','GT32','GT35','GT47','GT51',
          'GH1','GH2','GH3','GH4','GH13','GH18','GH23','GH25','GH28','GH31','GH32','GH43','GH55',
          'GH66','GH68','GH72','GH73','GH77','CE1','CE4','CE9','CE11','CBM34','CBM41','CBM50']
cazy_stats = np.zeros((2,len(groups)) )
for i in range(len(groups) ):
    cazy_stats[0][i] = list(st_anno_table['group']).count(groups[i])
    cazy_stats[1][i] = list(lb_anno_table['group']).count(groups[i])
    
```

```python
fig,ax = plt.subplots(1,1,figsize=(20,1))
ax = sns.heatmap( cazy_stats,annot=True,xticklabels = groups,yticklabels=['ST','LB'],\
                 cmap=sns.diverging_palette(220, 20, as_cmap=True) )
ax.set_xlabel('Enzyme classes')
ax.set_ylabel('Species')
ax.set_title('Functional annotation of carbonhydrate active enzymes')
plt.savefig('../../Figures/cazy_annot.svg',format='svg',dpi=150)
```

Compared to LB, ST's genome is enriched with CBM41(Carbohydrate-Binding Module Family 41),Modules of approx. 100 residues found in primarily in bacterial pullulanases. 

In general, more carbonhydrate ative enzymes are identified from ST's genome.
