# simulation & validation

```python
ic_ref = {'lb':0.09,'st':0.09,'lcts_e':191.44,'lac__L_e':0,'ac_e':0,'for_e':2,
          '3mba_e':0,'4hba_e':0,'gcald_e':0,'succ_e':0,'2mba_e':0,'2mpa_e':0}

for i in range(len(freeAA_mM['AA_ids'])):
    aa_id = list(freeAA_mM['AA_ids'])[i]+'_e'
    ic_ref[aa_id] = float( list( freeAA_mM['AVG(0)'] )[i] )
ic_ref['met__L_e'] = 0.00173
path_params = '../data/GSMMs/pa_params.csv'
ptot_st, ptot_lb = 0.46, 0.529
```

```python
st_params = get_pa_params( path_params, 'st' )
lb_params = get_pa_params( path_params, 'lb' )
```

```python
times,step = np.linspace(0,6,num= 1000,retstep=True)
profile = {key: [ic_ref[key]] for key in ic_ref.keys() }
st_f, lb_f = {'Growth':[],'EX_lcts_e':[]}, {'Growth':[],'EX_lcts_e':[],'EX_caspep_e':[] }
for k in ic_ref.keys():
    if ( ('EX_'+k) not in st_f) and (k not in ['st','lb']):
        st_f[ 'EX_'+k ], lb_f[ 'EX_'+k ] =[],[]

for i in range(len(times)-1):
    profile_t = {k: profile[k][i] for k in profile.keys() }#concentration profile snapshot at time t = i
    st_fluxes = run_pfba_pa( st, ptot_st, growth_funcs['st'], profile_t, st_params ) 
    lb_fluxes = run_pfba_pa( lb, ptot_lb, growth_funcs['lb'], profile_t, lb_params )
    for k in st_f.keys():
        st_f[k].append( st_fluxes[k] )
    for k in lb_f.keys():
        lb_f[k].append( lb_fluxes[k] )

    for k in profile.keys():
        if k == 'st':
            profile['st'].append( profile['st'][i] + st_f['Growth'][i]*profile['st'][i]*step )
        elif k == 'lb':
            profile['lb'].append( profile['lb'][i] + lb_f['Growth'][i]*profile['lb'][i]*step )
        else:
            profile[k].append( max( 0, profile[k][i] + lb_f[ 'EX_'+k ][i]*profile['lb'][i]*step +\
                              st_f[ 'EX_'+k ][i]*profile['st'][i]*step ) )#concentration >= 0
```
